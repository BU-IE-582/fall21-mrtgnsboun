```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv("Musk1.csv",header=None)
```

```python
data = data.rename(columns={0:"Class",1:"Id"})
```

```python
data.groupby("Id").describe()
```

### METHOD 1

```python
data = data.groupby("Id").max()
```

### CLASSIFIER 1

```python
X = data.drop(["Class"],axis=1)
y = data["Class"]
```

```python
mdl = RandomForestClassifier(n_estimators=100,max_depth=50)
mdl.fit(X,y)
print(np.mean(cross_val_score(mdl,X,y,cv = 10)))
```

### CLASSIFIER 2

```python
mdl = LogisticRegression(max_iter=1000)
mdl.fit(X,y)
print(np.mean(cross_val_score(mdl,X,y,cv = 10)))
```

### METHOD 2

```python
data = data.groupby("Id").min()
```

### CLASSIFIER 1

```python
X = data.drop(["Class"],axis=1)
y = data["Class"]
```

```python
mdl = RandomForestClassifier(n_estimators=100,max_depth=50)
mdl.fit(X,y)
print(np.mean(cross_val_score(mdl,X,y,cv = 10)))
```

### CLASSIFIER 2

```python
mdl = LogisticRegression(max_iter=1000)
mdl.fit(X,y)
print(np.mean(cross_val_score(mdl,X,y,cv = 10)))
```
