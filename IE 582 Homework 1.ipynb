```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance
from matplotlib import image
```

```python
x = []
num_area = 0

for a in range(1,16):
    for k in range(1000):
        n = np.random.uniform(-1,1,(1,a))
        dst = distance.euclidean(n,0)
        if dst <= 1:
            num_area = num_area+1
    a = num_area/1000
    x.append(a)
    num_area = 0

plt.xlabel("dimension")
plt.ylabel("values")
plt.plot(x)
```

```python
pi_2D = 4*x[1]
pi_3D = 6*x[2]
print(pi_2D,pi_3D,np.pi)
```

```python
iterations = [5000,10000,25000,50000,100000]
num_area2 = 0
num_area3 = 0
pi2 = []
pi3 = []

for i in range(5):
    for k in range(iterations[i]):
        n2 = np.random.uniform(-1,1,(1,2))
        n3 = np.random.uniform(-1,1,(1,3))
        dst2 = distance.euclidean(n2,0)
        dst3 = distance.euclidean(n3,0)
        if dst2 < 1:
            num_area2 = num_area2+1
        if dst3 < 1:
            num_area3 = num_area3+1
    a2 = num_area2/iterations[i]
    a3 = num_area3/iterations[i]
    pi_2D = 4*a2
    pi_3D = 6*a3
    pi2.append(pi_2D)
    pi3.append(pi_3D)
    num_area2 = 0
    num_area3 = 0
```

```python
pi = [np.pi,np.pi,np.pi,np.pi,np.pi]
```

```python
plt.ylim(3.1,3.2,0.1)
plt.plot(iterations,pi2)
plt.plot(iterations,pi3)
plt.plot(iterations,pi)
```

It can be said that as the number of iterations increases, the simulation result converges to pi more.

```python
num_area = 0
averages = []

for a in range(1,16):
    for k in range(1000):
        min_values = []
        n = np.random.uniform(-1,1,(1,a))
        additional = []
        for b in range(100):
            l = np.random.uniform(-1,1,(1,a))
            additional.append(l)
        distances = []
        for c in range(100):
            dst = distance.euclidean(n,additional[c])
            distances.append(dst)
        min_values.append(min(distances))
    averages.append(min_values)
```

```python
plt.plot(averages)
```
